```ruby
require "rbplotly"
include Math
nil
```

```ruby
def _fn
    xs = []
    ys = []
    x = 0
    while x < Math::PI * 2.1
        xs << x
        ys << yield(x)
        x += 0.05
    end
    { x: xs, y: ys }
end

def fn(f, name: nil)
  trace = _fn{ |x| f.call(x) }
  trace[:name] = name || f.inspect
  trace
end

def plot(data)
   Plotly::Plot.new(data: data).show 
end
```

```ruby
data = []
data << fn(->(x){ sin(x) }, name: "sin(x)")
data << fn(->(x){ sin(x) * sin(x) }, name: "sin(x) * sin(x)")
plot data
```
